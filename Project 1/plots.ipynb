```python
## IMPORTS
import xarray as xr
import earthkit.data as ekd
```

```python
## BOUNDING BOXES
# go to gmaps, pick top-left, bottom-right

Fuzhou_BB = [26.1, 119.3, 26.1, 119.3]
Fujian_BB = [31.7, 115.0, 23.3, 123.2] ## covers fujian, zhejiang, southmost bit of anhui, bit of taiwan

SCSea_BB = [21.8, 105.4, 8.9, 121.6]
IOcean_BB = [25.9, 41.6, -15.7, 101.0]
WPacific_BB = [20.0, 122.1, -32.0, -72.4] # focus on pressure
```

```python
## BASE REQUEST TEMPLATE

years = [
    str(yr) for yr in range(1940, 2025)
]

months = [
    '01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12'
]

REQUEST_TEMPLATE = {
        'product_type': 'monthly_averaged_reanalysis',
        'year': years,
        'month': months,
        'time': ['00:00'],
        'data_format': 'netcdf',
        'download_format': 'unarchived'
}
```

## requests
close section when not in use

```python
# FUZHOU REQUESTS

request = REQUEST_TEMPLATE
request['area'] = Fuzhou_BB
#request['variable'] = '2m_temperature'
#request['variable'] = ["10m_u_component_of_wind", "10m_v_component_of_wind"]
#request['variable'] = '2m_dewpoint_temperature'

data = ekd.from_source('cds', 'reanalysis-era5-single-levels-monthly-means', request)
```

```python
# FUJIAN REQUESTS

request = REQUEST_TEMPLATE
request['area'] = Fujian_BB
#request['variable'] = '2m_temperature'
#request['variable'] = ["10m_u_component_of_wind", "10m_v_component_of_wind"]
#request['variable'] = '2m_dewpoint_temperature'
#request['variable'] = 'mean_sea_level_pressure'

data = ekd.from_source('cds', 'reanalysis-era5-single-levels-monthly-means', request)
```
